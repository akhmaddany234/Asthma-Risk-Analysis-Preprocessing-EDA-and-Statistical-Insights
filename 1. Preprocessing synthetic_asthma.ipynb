2.1 Mengatasi missing value pada kolom 'Allergies' menggunakan modus


```python
from sklearn.impute import SimpleImputer

imputer_mode = SimpleImputer(strategy="most_frequent")

data["Allergies"] = imputer_mode.fit_transform(data[["Allergies"]]).ravel()
```

*Alasan:* Untuk kolom Allergies yang memiliki missing value sekitar 29%, pendekatan yang sesuai untuk mengatasi masalah ini kita lakukan imputasi menggunakan modus. Hal ini karena variabel kolom ini bersifat kategorikal, sehingga metode ini seperti mean atau median tidak relevan. Dengan kita memilih modus, kita dapat mengisi nilai kosong berdasarkan kategori yang paling sering muncul pada kolom tersebut


2.2 Mengatasi missing value pada kolom 'Comorbidities' menggunakan modus

```python
mode_val = data["Comorbidities"].mode()[0]
data["Comorbidities"] = data["Comorbidities"].fillna(mode_val)
```

*Alasan:* Dari penjelasan mentor, jika missing value <5% maka data dapat dihapus karena jumlahnya kecil dan tidak terlalu berpengaruh. Sedangkan jika missing value â‰¥5%, maka data diimputasi agar tetap dapat digunakan, kali ini teknik imputasi yang dipilih adalah modus karena tipe data kolom "Comorbidities" kategorikal. Namun, sebenarnya cara mengatasi missing value juga bergantung pada jenis analisis yang dilakukan. Beberapa jurnal menyebutkan bahwa apabila missing value lebih dari 50%, sebaiknya kolom tersebut dipertimbangkan untuk dihapus, karena imputasi pada kondisi ini berisiko menghasilkan data yang bias dan tidak akurat akibat terlalu banyak nilai tebakan.

2.3 Mengatasi missing value pada kolom "Asthma_Control_Levels" dengan 'Unknown'

```python
data['Asthma_Control_Level'] = data['Asthma_Control_Level'].fillna("Unknown")
```

Kolom Asthma_Control_Level merupakan data kategorikal sehingga cocok jika missing value diatasi dengan menambahkan label baru "Unknown". Hal ini dipilih karena apabila langsung diimputasi menggunakan modus, dapat menambah dominasi salah satu kategori secara tidak wajar, mengingat persentase missing value pada kolom ini lebih dari 50%.

Selain itu, berdasarkan jurnal atau dosen saya terkait cleaning data, jika suatu kolom memiliki missing value lebih dari 50% maka kolom tersebut biasanya lebih dipertimbangkan untuk dihapus karena dikhawatirkan tidak representatif. Namun, sesuai penjelasan dari mentor sebelumnya, data dengan missing value lebih dari 5% masih dapat ditangani dengan teknik imputasi. Oleh karena itu, pada kasus ini missing value diatasi dengan menambahkan kategori baru "Unknown", agar data tetap dapat digunakan tanpa menimbulkan bias berlebihan atau sekadar menebak nilai yang hilang.

## **4. Cek Outlier** (Ciara, Fuzi)

```python
import numpy as np

# Pilih kolom numerik
numeric_cols = data.select_dtypes(include=[np.number]).columns

# Dictionary untuk simpan hasil
outlier_counts = {}

for col in numeric_cols:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    outlier_counts[col] = outliers.shape[0]

outlier_counts
```

Berdasarkan hasil deteksi outlier menggunakan metode Interquartile Range (IQR), diketahui bahwa variabel Age, Medication Adherence, dan Peak Expiratory Flow tidak memiliki outlier sehingga distribusi datanya relatif normal dan tidak terdapat nilai ekstrem yang signifikan. Sementara itu, pada variabel BMI ditemukan sebanyak 24 outlier, yang mengindikasikan adanya individu dengan indeks massa tubuh jauh lebih tinggi atau lebih rendah dibanding kelompok mayoritas. Variabel Number of ER Visits memiliki 3 outlier, yang menunjukkan adanya pasien dengan jumlah kunjungan Unit Gawat Darurat yang tidak wajar dibanding pasien lainnya. Adapun variabel dengan jumlah outlier terbanyak adalah FeNO Level, yaitu sebanyak 41, yang menandakan adanya variasi fisiologis ekstrem atau kemungkinan kesalahan pencatatan data. Dengan demikian, variabel yang perlu mendapat perhatian khusus dalam analisis selanjutnya adalah BMI dan FeNO Level, karena keberadaan outlier dalam jumlah besar dapat memengaruhi hasil analisis statistik maupun pemodelan yang akan dilakukan.
